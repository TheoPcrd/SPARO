# Prediction 2000 / 2020

```python
from datetime import datetime
from datetime import timedelta

dir_save_supermodel = '/home/datawork-lemar-apero/tpicard/DATA_SAT/2000_2023/prediction/'
name_nc = 'supermodel_ssh80_sst24_w100.nc'

nc_file = nc.Dataset(dir_save_supermodel+name_nc,'r')
prediction_pdf = nc_file.variables["supermodel"][:]
mid_date_pdf = nc_file.variables["mid_date"][:]
std = nc_file.variables["std"][:]
nc_file.close()

# Path st data

nan_threshold = 1
file_st_data = "/home/datawork-lemar-apero/tpicard/DATA_ST/PAP_data_3000m 1989_2019_theo.xlsx"

#%%% Chloro data

file_satellite_chloro = "/home/datawork-lemar-apero/tpicard/DATA_SAT/2000_2023/chl_sat_tint.nc"
nc_file = nc.Dataset(file_satellite_chloro,'r')
chloro_images = nc_file.variables["chl"][::10]
mid_date_chloro = nc_file.variables["time"][::10]
nc_file.close()

mid_date_chloro_datetime=[]
for date in mid_date_chloro:
    mid_date_chloro_datetime.append(datetime.fromtimestamp(date))




gap = 38 # Mean box 200km

time_serie_chloro = np.nanmean(chloro_images[:,gap:-gap,gap:-gap],axis=(1,2))
nan_coef_box = np.sum(np.where(np.isnan(chloro_images[:,gap:-gap,gap:-gap]),1,0),axis=(1,2)) / (chloro_images[0,gap:-gap,gap:-gap].flatten().shape[0])
time_serie_chloro = np.ma.masked_where(nan_coef_box > nan_threshold, time_serie_chloro)

box_size = (100 - gap*2)*8

#mid_date_chloro_datetime = [datetime.strptime(date, '%Y-%m-%d').date() for date in mid_date_chloro]
first_date_chloro_datetime = [(date+timedelta(days=-5)) for date in mid_date_chloro_datetime]
last_date_chloro_datetime = [(date+timedelta(days=5)) for date in mid_date_chloro_datetime]
```

```python
# Choose the date

last_year = 2019
first_year = 2000

# Lire le fichier Excel
df = pd.read_excel(file_st_data)
# Cr√©er des variables pour chaque colonne du tableau
mid_date = df['mid date (dd/mm/yyyy)']
dry_weight = df['Dry Weight Flux mg/m2/day']

#%%% Choose date
df_filtre = df.loc[df['mid date (dd/mm/yyyy)'].dt.year <= last_year]
df_filtre = df_filtre.loc[df['mid date (dd/mm/yyyy)'].dt.year >= first_year]

mid_date = df_filtre['mid date (dd/mm/yyyy)']
last_date = df_filtre['close date (dd/mm/yyyy)']
first_date = df_filtre['open date (dd/mm/yyyy)']
first_day = df_filtre['open day julian']
last_day = df_filtre['close day julian']
delta_day = first_day-last_day

dry_weight = df_filtre['Dry Weight Flux mg/m2/day']
POC = df_filtre['Particulate Organic Carbon Flux mg/m2/day']
PIC = df_filtre['Particulate Inorganic Carbon Flux mg/m2/day']
TPN = df_filtre['Total Particulate Nitrogen Flux mg/m2/day']
```

```python
import warnings
warnings.filterwarnings('ignore')

fig = plt.figure(figsize=(20,20))
    
size_marker = 20

ax1 = fig.add_subplot(3,1,1)

plt.axvline(datetime(2009,1,1),color='grey',linestyle='--')
plt.axvspan(datetime(2000,1,1), datetime(2009,1,1), alpha=0.2, color='grey',zorder=-1)


label = 'a'
trans = mtransforms.ScaledTranslation(10/72, -5/72, fig.dpi_scale_trans)
ax1.text(0.96, 0.95, label, transform=ax1.transAxes + trans,
        fontsize=20, verticalalignment='top',
        bbox=dict(boxstyle='round', facecolor='white', alpha=1, pad=0.2))

for i,j in zip(mid_date.index[:-1],mid_date.index[1:]):

    #print(first_date[i])
    if np.isnan(POC[i])==False:
        #print(i)
        if i != 269:
            ax1.plot([first_date[i],last_date[i]],[dry_weight[i],dry_weight[i]],'-',color='k',alpha=0.8,linewidth=3)
            if last_date[i]!=first_date[j]:
                ax1.scatter(last_date[i],dry_weight[i],marker='^',color='k',s=size_marker)
                ax1.scatter(first_date[i],dry_weight[i],marker='^',color='k',s=size_marker)
            elif i != 268:
                ax1.plot([last_date[i],first_date[j]],[dry_weight[i],dry_weight[j]],'-',color='k',alpha=0.8,linewidth=3)
                ax1.scatter(last_date[i],dry_weight[i],marker='^',color='k',s=size_marker)
                ax1.scatter(first_date[i],dry_weight[i],marker='^',color='k',s=size_marker)
            
#ax1.scatter(last_date[i],dry_weight[i],marker='x',color='k',s=size_marker,label="start sampling")
#ax1.scatter(first_date[i],dry_weight[i],marker='^',color='k',s=size_marker,label="end sampling")
plt.ylim(bottom=0)

ax1_bis=ax1.twinx()
ax1_bis.plot(mid_date_chloro_datetime,time_serie_chloro,'-',color='g',alpha=0.4,linewidth=3,zorder=0)
ax1_bis.fill_between(mid_date_chloro_datetime,0,time_serie_chloro,'-',color='g',alpha=0.2,linewidth=3,zorder=0)        
plt.ylim(0,2.5)

#plt.legend(loc=2)
ax1.set_ylabel('Dry weight \n [$mg.m^{-2}.day^{-1}$]',color='k')   
plt.grid()
plt.xlim(datetime(2000,1,1),datetime(2020,1,1))
#plt.xticks([])

ax2 = fig.add_subplot(3,1,2)

plt.axvline(datetime(2009,1,1),color='grey',linestyle='--')
plt.axvspan(datetime(2000,1,1), datetime(2009,1,1), alpha=0.2, color='grey',zorder=-1)



label = 'b'
trans = mtransforms.ScaledTranslation(10/72, -5/72, fig.dpi_scale_trans)
ax2.text(0.96, 0.95, label, transform=ax2.transAxes + trans,
        fontsize=20, verticalalignment='top',
        bbox=dict(boxstyle='round', facecolor='white', alpha=1, pad=0.2))

for i,j in zip(mid_date.index[:-1],mid_date.index[1:]):

    #print(first_date[i])
    if np.isnan(POC[i])==False:
        #print(i)
        if i != 269:
            ax2.plot([first_date[i],last_date[i]],[POC[i],POC[i]],'-',color='k',alpha=0.8,linewidth=3)
            if last_date[i]!=first_date[j]:
                ax2.scatter(last_date[i],POC[i],marker='^',color='k',s=size_marker)
                ax2.scatter(first_date[i],POC[i],marker='^',color='k',s=size_marker)
            elif i != 268:
                ax2.plot([last_date[i],first_date[j]],[POC[i],POC[j]],'-',color='k',alpha=0.8,linewidth=3)
                ax2.scatter(last_date[i],POC[i],marker='^',color='k',s=size_marker)
                ax2.scatter(first_date[i],POC[i],marker='^',color='k',s=size_marker)
plt.ylim(bottom=0)
#plt.legend(loc=2)
ax2.set_ylabel('POC \n [$mg.m^{-2}.day^{-1}$]',color='k')   
  
ax2_bis=ax2.twinx()
ax2_bis.plot(mid_date_chloro_datetime,time_serie_chloro,'-',color='g',alpha=0.4,linewidth=3,zorder=0)
ax2_bis.fill_between(mid_date_chloro_datetime,0,time_serie_chloro,'-',color='g',alpha=0.2,linewidth=3,zorder=0)        
plt.ylim(0,2.5)
plt.ylabel("Chl [$mg.m^{-3}$]",color='g')

plt.grid()
plt.xlim(datetime(2000,1,1),datetime(2020,1,1))
#plt.xticks([])

ax4 = fig.add_subplot(3,1,3)

plt.axvline(datetime(2009,1,1),color='grey',linestyle='--')
plt.axvspan(datetime(2000,1,1), datetime(2009,1,1), alpha=0.2, color='grey',zorder=-1)



label = 'c'
trans = mtransforms.ScaledTranslation(10/72, -5/72, fig.dpi_scale_trans)
ax4.text(0.96, 0.95, label, transform=ax4.transAxes + trans,
        fontsize=20, verticalalignment='top',
        bbox=dict(boxstyle='round', facecolor='white', alpha=1, pad=0.2))

for i,j in zip(mid_date.index[:-1],mid_date.index[1:]):

    #print(first_date[i])
    if np.isnan(POC[i])==False:
        #print(i)
        if i != 269:
            ax4.plot([first_date[i],last_date[i]],[PIC[i],PIC[i]],'-',color='k',alpha=0.8,linewidth=3)

            if last_date[i]!=first_date[j]:
                ax4.scatter(last_date[i],PIC[i],marker='^',color='k',s=size_marker)
                ax4.scatter(first_date[i],PIC[i],marker='^',color='k',s=size_marker)
                
            elif i != 268:

                ax4.plot([last_date[i],first_date[j]],[PIC[i],PIC[j]],'-',color='k',alpha=0.8,linewidth=3)
                ax4.scatter(last_date[i],PIC[i],marker='^',color='k',s=size_marker)
                ax4.scatter(first_date[i],PIC[i],marker='^',color='k',s=size_marker)

plt.ylim(bottom=0)
#plt.legend(loc=2)
ax4.set_ylabel('PIC \n [$mg.m^{-2}.day^{-1}$]',color='k')   
ax4_bis=ax4.twinx()
ax4_bis.plot(mid_date_chloro_datetime,time_serie_chloro,'-',color='g',alpha=0.4,linewidth=3,zorder=1)
ax4_bis.fill_between(mid_date_chloro_datetime,0,time_serie_chloro,'-',color='g',alpha=0.2,linewidth=3,zorder=1)        


plt.ylim(0,2.5)
plt.grid()
plt.xlim(datetime(2000,1,1),datetime(2020,1,1))


plt.savefig("ST_time_serie_2000_2020.png",dpi=300,bbox_inches='tight',facecolor="white")
```
